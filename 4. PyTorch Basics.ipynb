```python
a = torch.tensor([1,2,3])
b = torch.tensor([7,5,3])
```

```python
a + b == a.add(b)
```

```python
a * b == a.mul(b)
```

a.add(b) doesn't change *a*

a.add_(b) changes *b*

```python
a * b
a.mul(b)
```

```python
a.dot(b)
```

```python
a = torch.tensor([[1,2,3],[4,5,6]])
b = torch.tensor([[1,2],[3,4],[5,6]])
```

```python
a.mm(b)
```

```python
a @ b
```

.mul() and * means elementwise multiplication

.dot() is the dot product

.mm() and @ means matrix multiplication
